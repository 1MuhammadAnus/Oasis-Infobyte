# **Unemployment Analysis with Python**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly .express as px
```

# **Load Dataset**

```python
df = pd.read_csv("/content/Unemployment in India.csv")
df.head(10)
```

# **Renmeing Columns**

```python
df.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate',
              'Estimated Employed', 'Estimated Labour Participation Rate', 'Region']
```

```python
df.isnull()
```

# **Estimated Unemployment Rate Over Time**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate', data=df, marker='o')
plt.title('Estimated Unemployment Rate Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns=['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=df)
plt.show()
```

# **Indian Unemployment**

# **CHECKING THE CORRELATION BETWEEN THE FEATURE OF DATASET**

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(8, 6))
ax = sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
# ax = sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

# **Dashboard to analyze the unemployment rate**

```python
unemployment_grouped = df.groupby(['Region', 'States']).agg({'Estimated Unemployment Rate': 'mean'}).reset_index()

# Now, plot the sunburst chart
figure = px.sunburst(unemployment_grouped,
                     path=['Region', 'States'],
                     values='Estimated Unemployment Rate',
                     width=700, height=600,
                     color_continuous_scale='RdYlGn',
                     title="Unemployment Rate in India")
figure.show()
```
